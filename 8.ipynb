```python
import numpy as np # Th vin h tr tnh ton
import pandas as pd # Th vin h tr x l d liu dng bng
from sklearn.cluster import KMeans # Th vin h tr phn cm
import matplotlib.pyplot as plt # Th vin h tr v  th
import seaborn as sns # Th vin h tr v  th

# To d liu gi nh
np.random.seed(42) #  kt qu gi nguyn qua cc ln chy
n_samples = 200 # S lng mu d liu
frequency = np.random.randint(1, 10, n_samples)  # Tn sut mua hng
avg_order_value = np.random.randint(10, 100, n_samples)  # Gi tr trung bnh ca n hng
num_products = np.random.randint(1, 5, n_samples)  # S lng sn phm trong gi

# To DataFrame
data = pd.DataFrame({ # To DataFrame t dict d liu
    'Frequency': frequency, # Tn sut mua hng
    'AvgOrderValue': avg_order_value, # Gi tr trung bnh ca n hng
    'NumProducts': num_products # S lng sn phm trong gi
})

# S dng KMeans  phn cm (gi s 3 cm)
kmeans = KMeans(n_clusters=3, random_state=42) # Khi to m hnh KMeans vi 3 cm
data['Cluster'] = kmeans.fit_predict(data) # Gn nhn cm cho tng mu d liu

# Hin th cc trung tm cm
centers = kmeans.cluster_centers_ # Trung tm ca cc cm (3, 2)

# V biu  phn cm
plt.figure(figsize=(10, 6)) # Kch thc biu 
sns.scatterplot(data=data, x='Frequency', y='AvgOrderValue', hue='Cluster', palette='viridis', s=100) # V biu  phn cm
plt.scatter(centers[:, 0], centers[:, 1], s=300, c='red', label='Cluster Centers') # V trung tm cm
plt.title("Phn cm khch hng da trn hnh vi mua hng") # Tiu  biu 
plt.xlabel("Tn sut mua hng") # Tn trc tung
plt.ylabel("Gi tr trung bnh ca n hng") # Tn trc honh
plt.legend() # Hin th ch thch
plt.show() # Hin th biu 
```
