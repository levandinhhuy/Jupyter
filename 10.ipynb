```python
import matplotlib.pyplot as plt # Th vin v  th matplotlib (cn ci t: pip install matplotlib)
from sklearn.decomposition import PCA # Th vin gim chiu d liu PCA ca scikit-learn
from sklearn.datasets import load_digits # D liu hnh nh ch vit tay (0-9)
import seaborn as sns # Th vin v  th p mt (cn ci t: pip install seaborn)

# Ti d liu hnh nh ch vit tay (0-9)
digits = load_digits() # D liu gm 1797 nh, mi nh c kch thc 8x8 pixels
data = digits.data # 1797 nh c biu din di dng ma trn 64 chiu (8x8)
target = digits.target # Nhn ca nh (0-9) tng ng vi mi nh

# p dng PCA  gim chiu t 64 xung cn 2
pca = PCA(n_components=2) # Gim xung cn 2 chiu (PC1, PC2)  v  th
data_pca = pca.fit_transform(data) # Thc hin gim chiu d liu v chuyn i d liu
 
# Chuyn kt qu thnh DataFrame  trc quan ha
import pandas as pd # Th vin x l d liu dng bng (cn ci t: pip install pandas)
df = pd.DataFrame(data_pca, columns=['PC1', 'PC2']) # Chuyn kt qu thnh DataFrame
df['Target'] = target # Thm ct nhn (target) vo DataFrame

# V biu  phn tn ca d liu sau khi gim chiu
plt.figure(figsize=(10, 8)) # To kch thc biu  rng 10 inch, cao 8 inch
sns.scatterplot(data=df, x='PC1', y='PC2', hue='Target', palette='tab10', s=60, legend='full') # V biu  phn tn vi 10 mu khc nhau
plt.title("PCA of Handwritten Digits Dataset (2 Components)") # t tiu  biu  l "PCA of Handwritten Digits Dataset (2 Components)"
plt.xlabel("Principal Component 1") # t tn trc honh l "Principal Component 1" (PC1)   
plt.ylabel("Principal Component 2") # t tn trc tung l "Principal Component 2" (PC2)
plt.show() # Hin th biu  va v 
```
